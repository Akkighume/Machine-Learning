```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita_income_(US$)')
plt.scatter(df['year'], df['per_capita_income_(US$)'], color='green', marker='+')
```

```python
new_df = df.drop('per_capita_income_(US$)',axis='columns')
new_df
```

```python
# Create Linear Regression object
reg = linear_model.LinearRegression()
reg.fit(new_df, df['per_capita_income_(US$)'])
```

**(1)Predict canada's per capita income in year 2020**

```python
reg.predict([[2020]])   # x-year
```

```python
reg.coef_  # m-slope
```

```python
reg.intercept_   # b-intercept
```

**Y = m * X + b (m is coefficient and b is intercept)**
